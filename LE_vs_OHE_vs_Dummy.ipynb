```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
cars = pd.read_csv("D:\\WERK\\PYTHON\\PORTO\\CONTENT\\3. LABEL ENCODER VS ONE HOT VS DUMMY\\car_dataset.csv")
cars
```

```python
le = LabelEncoder()
df_le = cars
df_le['Level of Car'] = le.fit_transform(df_le['Level of Car'])
df_le
```

```python
ohe = OneHotEncoder()
df_ohe = df_le
df_ohe = ohe.fit_transform(df_le[['Car Model']])
df_ohe = pd.DataFrame(df_ohe.toarray(), columns=ohe.get_feature_names_out(['Car Model']))
df_merged_ohe = pd.concat([df_le, df_ohe], axis=1)
```

```python
df_merged_ohe
```

```python
df_dummies = df_le
df_dummies = pd.get_dummies(df_dummies['Car Model'])
df_merged_dummies = pd.concat([df_le, df_dummies], axis=1)
df_merged_dummies
```
